```python
df = pd.DataFrame(data)
```

```python
more_than_3 = df[df['Customer Satisfaction'] > 3].shape[0]
less_than_or_equal_to_3 = df[df['Customer Satisfaction'] <= 3].shape[0]
```

```python
labels = ['More than 3', 'Less than or equal to 3']
sizes = [more_than_3, less_than_or_equal_to_3]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Customer Satisfaction Ratings')
plt.show()
```

```python
# Count occurrences where customer satisfaction is more than 3 with special offers
with_offers = df[(df['Customer Satisfaction'] > 3) & (df['Special Offers'] == 'Yes')].shape[0]

# Count occurrences where customer satisfaction is more than 3 without special offers
without_offers = df[(df['Customer Satisfaction'] > 3) & (df['Special Offers'] == 'No')].shape[0]

# Plotting the bar chart
plt.bar(['With Special Offers', 'Without Special Offers'], [with_offers, without_offers] , color = 'orange')
plt.xlabel('Special Offers')
plt.ylabel('Count')
plt.title('Customer Satisfaction Ratings > 3 with and without Special Offers')
plt.show()
```

```python
max_sales_index = df['Total'].idxmax()

# Retrieve the row with the highest total sales
row_with_max_sales = df.loc[max_sales_index]

print("Row with the highest sales:\n", row_with_max_sales)
```

```python
low_sales_index = df['Total'].idxmin()

# Retrieve the row with the highest total sales
row_with_low_sales = df.loc[low_sales_index]

print("Row with the lowest sales:\n", row_with_low_sales)
```
